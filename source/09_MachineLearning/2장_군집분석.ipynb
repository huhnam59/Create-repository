```python
#     
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=20, #  
                          n_features=2,  #  ( 20)
                          n_informative=2, #     
                          n_redundant=0,   #      
                          n_clusters_per_class=1, #     
                          n_classes=2, #   
                          random_state=123)
import numpy as np
print(X.shape, y.shape)
np.c_[X, y][:3]
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,3))
plt.scatter(x = X[:, 0], y=X[:,1], c=y, edgecolors='k')
plt.show()
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=2, #  
              init='k-means++', #    . random()
              n_init=10,  #    10      
              max_iter=300) #   300   
model.fit(X)
```

```python
centers = model.cluster_centers_ #   
print(centers)
print('model.labels :', model.labels_)
pred = model.predict(X)
print('       :', pred)
print('       :', y)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,3))
plt.scatter(x = X[:, 0], y=X[:,1], c=pred, edgecolors='k')
plt.scatter(x = centers[:, 0], y=centers[:, 1], c='r', marker='x', s=100, lw=5)
plt.show()
```

```python
from seaborn import load_dataset
iris = load_dataset('iris')
iris.sample(1)
```

```python
#      
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
iris['species'] = le.fit_transform(iris.species)
iris.loc[::50]
```

```python
from scipy.cluster.hierarchy import linkage
# method="single" :    
# method="complete" :    
cluster_model = linkage(iris, method="complete")
cluster_model.shape
```

```python
# dendrogram
from scipy.cluster.hierarchy import dendrogram
import matplotlib.pyplot as plt
plt.figure(figsize=(12,4))
dendrogram(cluster_model, labels=iris.index)
#  
plt.savefig('data/ch02_dendrogram.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 
from scipy.cluster.hierarchy import fcluster
fcluster(cluster_model, # linkage()   
        t=6, #   (threshold)
        criterion='distance')
```

```python
pred = fcluster(cluster_model, 4, criterion='distance')
pred
```

```python
# pred 1 0, 3 1, 2 2 
import numpy as np
np.choose([3, 3, 2, 1, 2], [0, 10, 20, 30]) # np.choose(, )
```

```python
# pred 1 0, 3 1, 2 2 
adjusted_pred = np.choose(pred, [999, 0, 2, 1]) #  (:iris.species)
import pandas as pd
pd.crosstab(iris.species, adjusted_pred, rownames=[''], colnames=[''],
           margins=True) #    
```

```python
original_name = le.inverse_transform(iris.species)
pred_name = le.inverse_transform(adjusted_pred)
pd.crosstab(original_name, pred_name)
```

```python
import seaborn as sns
sns.scatterplot(data=iris,
               x='petal_length', y='petal_width',
               hue=pred_name)
```

```python
np.random.seed(0)
X1 = np.random.rand(200)*3 # 0<= rand( )*3 < 3.
noise = np.random.normal(0, 0.2, X1.shape)
Y1 = np.sin(X1)+noise
#plt.scatter(x=X1, y=Y1)
group1 = np.c_[X1, Y1, np.full(len(X1), 0)]

X2 = X1 +  1.5
noise = np.random.normal(0, 0.2, X1.shape)
Y2 = np.cos(X2) + noise
#plt.scatter(X2, Y2)
group2 = np.c_[X2, Y2, np.full(len(X2), 1)]

data = np.r_[group1, group2]
plt.scatter(x=data[:, 0],
            y=data[:, 1], s=100, edgecolors='k')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
range_n_clusters = [2, 3, 4, 5, 6, 7, 8, 9, 10]
for n_cluester in range_n_clusters:
    model = KMeans(n_clusters=n_cluester,
                  n_init=10,
                  random_state=0
                  )
    model.fit(data[:, :-1])
    score = silhouette_score(X=data[:, :-1], labels=model.labels_)
    print("  : {},   :{}".format(n_cluester, score) )
#     7   
```

```python
model = KMeans(n_clusters=7, init='random', n_init=10)
model.fit(data[:, :-1])
pred = model.labels_
pred = model.predict(data[:, :-1])
```

```python
centers = model.cluster_centers_
colors = ['r', 'g', 'b', 'c', 'm', 'y', 'k'] # 0,1,2 => 0 / 3,4,5,6=>1
markers = ['>', 's', 'v', '^', 'o', '<', 'x']
# 0  
#data[pred==6]
for i, c in enumerate(centers):
    # print(i, c)
    plt.scatter(x=data[pred==i, 0],
                y=data[pred==i, 1], s=50, c=colors[i], marker=markers[i])
    plt.scatter(c[0], c[1], s=100, c='w', edgecolors='pink', lw=2)
```
