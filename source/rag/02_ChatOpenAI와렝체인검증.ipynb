```python
from openai import OpenAI
client = OpenAI()
response = client.chat.completions.create(
    model="gpt-4.1-nano",
    messages=[
        {"role":"system", "content":"너는 친절하게 대답해 주는 비서야"},
        {"role":"user",   "content":"2020년 월드 시리즈 누가 우승했어?"}
    ],
    temperature=0.5
)
```

```python
response.choices[0].message.content
```

```python
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(
  model = 'gpt-4.1-nano',
  temperature=0.5
)
ai_message = llm.invoke("2020년 월드 시리즈 누가 우승했어?")
```

```python
from langchain_openai import OpenAI
from langchain_core.messages import HumanMessage, SystemMessage
llm = ChatOpenAI(
    model="gpt-4.1-nano",  # 기본 모델: gpt-3.5-turbo
    temperature=1
)
# OpenAI SDK와 동일한 구조로 메시지 리스트 작성
messages = [
    # 시스템 메시지: AI의 역할 정의
    SystemMessage(content="너는 친절하게 대답해 주는 비서야"),
    # 사용자 메시지: 실제 질문
    HumanMessage(content="2020년 월드 시리즈 누가 우승했어?")
]
ai_message = llm.invoke(messages)
print(ai_message.content)
```
