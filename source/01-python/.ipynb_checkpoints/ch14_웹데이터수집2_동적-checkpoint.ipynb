```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')
result_list = soup.select('li > h3 > a')
for result in result_list:
    print("{} - {}".format(result.text, result.attrs['href']))
```

```python
result_list = soup.select('li > h3 > a')
for result in result_list:
    print("{} - {}".format(result.text, domain+result.attrs['href']))
```

## 2.2 페이징 처리
* 다음 뉴스 페이징 처리

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
driver=webdriver.Chrome()
driver.get('https://www.daum.net/')
# driver/implicitly_wait(0.5) # 0.5초 동안 대기
time.sleep(0.5) # 0.5초 멈춤
driver.find_element(By.CLASS_NAME, 'tf_keyword').click()
query=input('검색할 단어는?')
driver.find.element(By.CSS_SELECTOR, 'input[type="text"]'.send_keys(query))
driver.find_element(By.CLASS_NAME, 'btn_ksearch').click()
time.sleep(1)
# 뉴스탭 이동
driver.find_element(By.LINK,'뉴스').click()

# 뉴스 제목과 link를 list
news_list = []
pages=3 #
for page in range(1,pages+1):
    strongs = driver.find_elements(By.CSS_SELECTOR, 'div.item-title > strong')
    # print(len(strongs))
    for strong in strongs:
        a = strong.find_element(By.TAG_NAME, 'a')
        title = a.text
        link = a.get_attribute('href')
        news_list.append([title, link])
    page_nav=driver/find_elment(By.CSS_SELECTOR,'div.inner_paging')
    next_page=page_nav.find_elment(By.LINK_TEXT,str(page+1))
    next_page.clock()
    time.close()
driver.close()
import pandas as pd
pd.DataFrame(news_list,colums=['title','link'])
```

```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
driver=webdriver.Chrome()
driver.get('https://www.naver.com')
time.sleep(1)
elem=driver.find_element(By.ID,'query')
elem.send_keys('맞춤법 검사기')
elem.send_keys(Keys.RETURN) # form 태그의 submit 클릭 효과

textarea=driver.find_element(By.CLASS_NAME,'txt_gray')
results=''# 맞춤법 검사 완료된 text
for i, ready in enumerate(ready_text):
    print(f'검사중...{i+1}/{len(ready_text)}')
    #textarea.clear()
    textarea.send_keys(Keys.CONTROL,'a') # ctrl+a
    textarea.send_keys(ready)
    time.sleep(1)
    btn=driver.find_element(By.CLASS_NAME,'btn_check')
    btn.click()
    time.sleep(2)
    
    soup=BeautifulSoup(driver.page_source,'html.parser')
    result=soup.select_one('p._result_text.stand_txt').text
    results +=result +'\n'
    time.sleep(1)
```

```python
# 맞춤범 검사 결과를 파일로 출력
with open('data/ch14_맞춤법검사후.txt','w', encoding='utf-8') as f:
    f.write(results)
```

# 3절 연습문제
-영어번역 자동화 프로그램을 구현하시오. 네이버 맞춤법 검사를 마친 텍스트를 https://papago.naver.com/ 를 통해 영어로 번역된 텍스트를 파일로 출력하는 동적 웹크롤링 프로그램을 구현하시오.
제출 파일: 소스 html
         ch14_맞춤법검사후.txt
         ch14_자동화영어번역본.txt

```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

# 1. 맞춤법 검사 완료된 파일 읽기
with open('data/ch14_맞춤법검사후.txt', 'r', encoding='utf-8') as f:
    text = f.read()

# 2. Papago 접속
driver = webdriver.Chrome()
driver.get('https://papago.naver.com/')
time.sleep(2)  # 페이지 로딩 대기

# 3. 번역할 텍스트 준비 (Papago는 긴 텍스트도 처리 가능하지만 안정성을 위해 500자씩 나눔)
ready_text = []
temp_text = text
while len(temp_text) > 500:
    temp = temp_text[:500]
    last_dot_index = temp.rfind('\n')
    if last_dot_index == -1:
        last_dot_index = 500
    ready_text.append(temp_text[:last_dot_index])
    temp_text = temp_text[last_dot_index+1:]
ready_text.append(temp_text)

# 4. 번역 실행 및 결과 추출
translated_results = ''
for i, ready in enumerate(ready_text):
    print(f'번역 중... {i+1}/{len(ready_text)}')
    
    # 소스 텍스트 입력 영역 찾기
    textarea = driver.find_element(By.ID, 'txtSource')
    
    # 기존 텍스트 삭제 후 새 텍스트 입력
    textarea.send_keys(Keys.CONTROL, 'a')
    textarea.send_keys(Keys.DELETE)
    textarea.send_keys(ready)
    time.sleep(1)
    
    # 번역 버튼 클릭 (자동 번역이 안될 경우를 대비)
    try:
        btn_translate = driver.find_element(By.ID, 'btnTranslate')
        btn_translate.click()
    except:
        pass  # 자동 번역되는 경우 버튼 클릭 불필요
    
    time.sleep(3)  # 번역 완료 대기
    
    # 번역 결과 추출
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    result = soup.select_one('#txtTarget').text
    translated_results += result + '\n\n'
    
    time.sleep(1)

# 5. 번역 결과를 파일로 출력
with open('data/ch14_자동화영어번역본.txt', 'w', encoding='utf-8') as f:
    f.write(translated_results)

print("번역 완료! ch14_자동화영어번역본.txt 파일이 생성되었습니다.")
driver.close()
```
