# 12절 일관변경하기(결측치나 특정값)

## 12.1 fillna(결측치를 어떤 값으로 변경)
- 대부분 결측치" 중위수, 평균값, 최빈값, 이전값, 다음값. 
                 drop(dropna사용).그룹별 평균값, 그룹별 중위수(apply 함수 적용)로 대체

```python
df=pd.DataFrame([[np.nan,2,np.nan,0],
                [3,4,np.nan,1],
                [np.nan,3,np.nan,5],
                [np.nan,3,np.nan,4]],
                columns=list('ABCD'))
df
```

```python
# 모든 결측치를 무조건 0으로 대체
df.fillna(0)
```

```python
df
```

```python
# 결측치를 이전행 값으로 대체]
df.fillna(method='ffill')
```

```python
# 결측치를 다음행 값으로 대체
df.fillna(method='bfill')
```

```python
# 결측치를 열마다 특정 값으로 대체]
values={'A':99,'B':88, 'C':77, 'D':66}
df.fillna(value=values)
```

```python
df.mean(axis=0)
```

```python
# 결측치를 열평균으로 대체
df.fillna(value=df.mean(axis=0))
```

```python
# 결측치를 중위수(열중위값)로 대체
df.fillna(value=df.median(axis=0))
```

```python
# 결측치를 중위수(열중위값)로 대체
df.fillna(value=df.median(axis=0), inplace=True)
```

```python
df.mode().iloc[0] # 최빈값
```

```python
def fillna_grouped(series):
    'series의 결측치를 대체한 후 , series return'
    # pd.options.mode.copy_on_write=True
    seroes=series.copy()
    for col , item in series.items():
        if col!='species' and np.isnan(item): #결측치인지 여부
            species = series['species']
            series[col]=iris_mean.loc[species , col] #종별열별 평균값 대체
    return series
fillna_grouped(iris.iloc[3])
```

```python
iris.apply(fillna_grouped,axis=1)
```

```python
# 연습 8. iris 데이터의 각 열들 사이의 상관계수를 출력하세요.
import seaborn as sns
import pandas as pd

iris = sns.load_dataset('iris') # 1. iris 데이터 불러오기

corr_matrix = iris.corr(numeric_only=True) # 2. 상관계수 계산 (correlation matrix)

print("=== iris 데이터의 열 간 상관계수 ===") # 3. 결과 출력
print(corr_matrix)
```

```python
# 연습 9. iris 데이터에서 각 요소들과 변수들 평균과의 차이를 출력하세요. 
# 다음 그림은 처음 다섯 갱 행의 출력 결과입니다.
import seaborn as sns
import pandas as pd

iris = sns.load_dataset('iris') # 1. iris 데이터 불러오기

iris_numeric = iris.select_dtypes(include='number') # 2. 수치형 데이터만 선택

mean_values = iris_numeric.mean() # 3. 각 열의 평균 계산

diff_from_mean = iris_numeric - mean_values # 4. 각 요소와 평균의 차이 계산

print(diff_from_mean.head()) # 5. 결과 확인 (처음 5행)
```

```python
# 연습 10. iris  데이터의 종별 평균을 출력하세요.
import seaborn as sns
import pandas as pd

iris = sns.load_dataset('iris') # 1. iris 데이터 불러오기

species_mean = iris.groupby('species').mean(numeric_only=True) # 2. 종(species)별 평균 계산

print("=== iris 데이터의 종별 평균 ===") # 3. 결과 출력
print(species_mean)
```

```python
# 연습 11. iris 데이터에서 각 요소들과 종별 변수의 평균값의 차이를 출력하세요. 
# 각 종별로 3개씩 출력하세요. 다음 그림은 출력 예입니다
import seaborn as sns
import pandas as pd

iris = sns.load_dataset('iris') # 1. iris 데이터 불러오기

species_mean = iris.groupby('species').transform('mean') # 2. 종(species)별 평균 계산


diff_from_species_mean = iris.select_dtypes('number') - species_mean.select_dtypes('number')

print("=== setosa 종 중 3개 샘플 ===") # 4. 결과 확인: 각 종별로 3개씩 출력
print(diff_from_species_mean[iris['species'] == 'setosa'].head(3))

print("\n=== versicolor 종 중 3개 샘플 ===")
print(diff_from_species_mean[iris['species'] == 'versicolor'].head(3))

print("\n=== virginica 종 중 3개 샘플 ===")
print(diff_from_species_mean[iris['species'] == 'virginica'].head(3))
```
